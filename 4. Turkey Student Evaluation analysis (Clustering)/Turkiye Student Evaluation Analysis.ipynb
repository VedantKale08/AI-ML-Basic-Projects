# Principle Component Analysis

```python
#good technique for data reduction (dimentions i.e. columns)

X = df.iloc[:, 5:33]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=42) # n_components is no of dimentions we need to reduce
X_pca = pca.fit_transform(X)
```

```python
X_pca
```

```python
# how much we info retained from the dataset
# we need to retain 95% of original data for better processing
pca.explained_variance_ratio_.cumsum()[1]
```

# Model selection

```python
from sklearn.cluster import KMeans
distorations = []
cluster_range = range(1,6)

#elbow method to determine no of clusters
for i in cluster_range:
    model = KMeans(n_clusters=i, init='k-means++',random_state=42)
    model.fit(X_pca)
    distorations.append(model.inertia_)
```

```python
plt.plot(cluster_range, distorations)
plt.xlabel('No of clusters')
plt.ylabel('Distortions')
plt.show()
```

```python
# used best cluster
model = KMeans(n_clusters=3, init='k-means++',random_state=42)
model.fit(X_pca)
y = model.predict(X_pca)
```

```python
plt.scatter(X_pca[y==0, 0], X_pca[y==0, 1], s=50, c='red')
plt.scatter(X_pca[y==1, 0], X_pca[y==1, 1], s=50, c='yellow')
plt.scatter(X_pca[y==2, 0], X_pca[y==2, 1], s=50, c='green')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=100, c='blue')
plt.title("Cluster of students")
plt.xlabel('PCA 1')
plt.ylabel("PCA 2")
plt.legend()
plt.show()
```
